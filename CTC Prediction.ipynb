# Converting Objects to int & float

```python
df=df.replace({',': ''}, regex=True)
test=test.replace({',': ''}, regex=True)
```

```python
df['CTC']=df['CTC'].astype(np.float64)
df['Previous CTC']=df['Previous CTC'].astype(np.float64)

test['CTC']=df['CTC'].astype(np.float64)
test['Previous CTC']=df['Previous CTC'].astype(np.float64)
```

```python
df.info()
```

```python
col = ['College','Role','City type']

for c in col:
    le = LabelEncoder()
    df[c] = le.fit_transform(df[c])
    test[c] = le.fit_transform(test[c])
```

# Spliting into input & output

```python
Y_train = df['CTC']
del df['CTC']
X_train = df

Y_test = test['CTC']
del test['CTC']
X_test = test
```

# Training Model

```python
model = xgb.XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
pred = model.predict(X_test)
```

```python
metrics.r2_score(Y_train, pred)
```
